```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

```python
# Load the results
data = np.load("training_results_random_random.npz")
rewards = data["rewards"]
win_percentages = data["win_percentages"]

# Plot the rewards as a moving average
window_size = 1000
moving_avg_rewards = np.convolve(rewards, np.ones(window_size) / window_size, mode="valid")
plt.figure(figsize=(12, 6))
plt.plot(moving_avg_rewards, label="Moving Average Rewards")
plt.xlabel("Episodes")
plt.ylabel("Rewards")
plt.title("DQN Agent Training Rewards")
plt.legend()
plt.grid(True)
plt.show()
# Plot the win percentages
plt.figure(figsize=(12, 6))
plt.plot(win_percentages, label="Win Percentages")
plt.xlabel("Episodes")
plt.ylabel("Win Percentage")
plt.title("DQN Agent Win Percentages")
plt.legend()
plt.grid(True)
plt.show()


```
