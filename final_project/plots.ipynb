```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

```python
# Load the results
data = np.load("training_results_random_random.npz")
random_random_win_percentages = data["win_percentages"][0:50]

data = np.load("training_results_heuristic_random.npz")
heuristic_random_win_percentages = data["win_percentages"][0:50]

data=np.load("training_results_mixed_random.npz")
mixed_random_win_percentages = data["win_percentages"]

# Plot the win percentages as a moving average
sns.set(style="whitegrid")
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')
window_size = 10
random_random_moving_avg = moving_average(random_random_win_percentages, window_size)
heuristic_random_moving_avg = moving_average(heuristic_random_win_percentages, window_size)
mixed_random_moving_avg = moving_average(mixed_random_win_percentages, window_size)
plt.figure(figsize=(12, 6))
plt.plot(random_random_moving_avg, label='Trained with Random', color='blue')
plt.plot(heuristic_random_moving_avg, label='Trained with Heuristic', color='orange')
plt.plot(mixed_random_moving_avg, label='Trained with Mixed', color='green')
plt.title('Win Percentages (10-episode moving average) against Random Move Baseline')
plt.xlabel('Episodes (each episode is 1000 games)')
plt.ylabel('Win Percentage')
plt.legend()
plt.ylim(50,100)
plt.xlim(0, 40)


```
